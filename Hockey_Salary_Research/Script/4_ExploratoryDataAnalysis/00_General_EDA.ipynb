# General EDA of our merged Data

```python
import pandas as pd
import numpy as np
import dataprep as dp
from dataprep.eda import plot
import os

# Load in our data
filepath = '../../Data/entitiesResolved/merged_data.csv'
data = pd.read_csv(filepath)

data.head(-5)
```

```python
data.columns.to_list()
```

```python
# Change the name of column 'Y_SALARY_CAP' to 'SALARY_CAP' and 'Y_SALARY_CAP_PERCENTAGE' to 'SALARY_CAP_PERCENTAGE'
data.rename(columns={'Y_SALARY_CAP': 'SALARY CAP'}, inplace=True)
data.rename(columns={'Y_SALARY_CAP_PERCENTAGE': 'SALARY CAP PERCENTAGE'}, inplace=True)
```

Notice that we have a few similar column names: TEAM and AGE that are still overlapping from our merge. We should decide which of these to keep. 

## First, lets take a look at TEAM

```python
print(data['TEAM_x'].unique())
print(data['TEAM_x'].nunique())
print(data['TEAM_y'].unique())
print(data['TEAM_y'].nunique())
```

```python
num_records = data[data['TEAM_y'] == '-'].shape[0]
print(num_records)
```

Having 481 different values for TEAM seems unweildy, but there are 6858 records with '-' as the value for team, so let's keep the TEAM_x column and drop TEAM_y.

```python
data.drop(columns=['TEAM_y'], inplace=True)

# Rename the column 'TEAM_x' to 'TEAM'
data.rename(columns={'TEAM_x': 'TEAM'}, inplace=True)
data.columns.to_list()
```

# Now we should do the same thing with the AGE columns

```python
print(data['AGE_x'].unique())
print(data['AGE_x'].nunique())
print(data['AGE_y'].unique())
print(data['AGE_y'].nunique())
```

The values from our AGE_x columns unfortunately do not make a lot of sense, seeing as nobody who is 62 years old has ever played in the NHL. On the other hand,the values in our AGE_y column DO make a lot of sense, as there have been players who play and get paid up until they are in their mid forties, however it is very rare.

## How to proceed?

I think we should make a feature "Deceased" that extracts all rows where 'AGE_y' starts with 'Deceased', then convert 'AGE_y' to numeric, then replace 'AGE_x' with 'AGE_y', then finally rename 'AGE_x' to simply 'AGE'.

```python
# Create a new feature 'DECEASED' and one-hot-encode it
data['DECEASED'] = np.where(data['AGE_y'].str.startswith('Deceased'), 1, 0)

# Print the updated dataframe
print(data['DECEASED'].unique())
```

```python
# Extract the age from the 'AGE_y' column and convert it to an integer
data['AGE_y'] = data['AGE_y'].str.extract('(\d+)', expand=False).astype(int)
```

```python
data['AGE_x'] = data['AGE_y']
data.drop(columns=['AGE_y'], inplace=True)
data.rename(columns={'AGE_x': 'AGE'}, inplace=True)
```

```python
data.columns.to_list()
```

# Making sure columns are of the proper data type
Before delving too far into the analysis, we should make sure the data types of our columns are all appropriate (categorical types are categorical, and numerical types are numerical).

```python
# Select all columns that are non-numeric
non_numeric_columns = data.select_dtypes(exclude=[np.number]).columns.to_list()
non_numeric_columns
```

We can see that there are quite a few columns here that *should* be numerical. In fact, probably the majority of them. I think it may be easiest to simply manually separate them.

```python
# Select all columns from non_numeric_columns that contain '%' in the feature name
percentage_columns = [col for col in non_numeric_columns if '%' in col]

non_numeric_columns = list((set(non_numeric_columns) - set(percentage_columns)))

# Other numeric columns
numeric_columns = ['DRAFT ROUND', 'BASE SALARY', 'P.BONUS', 
                   'ROUND PICK', 'OVERALL DRAFT POSITION', 
                   'SALARY', 'PDO', 'AAV', 'IPP', 'CAP HIT', 
                   'S.BONUS']
```

```python
# Convert all percentage columns to numeric
for col in percentage_columns:
    # Replace '-' with NaN
    data[col] = data[col].replace('-', np.nan)
    data[col] = data[col].str.rstrip('%').astype(float) / 100.0

# view all percentage columns
data[percentage_columns].head()
```

```python
# Convert all columns that should be numeric to their proper data types
data['DRAFT ROUND'] = data['DRAFT ROUND'].replace('-', np.nan).astype('Int64')
data['BASE SALARY'] = data['BASE SALARY'].str.replace(',', '').str.replace('$', '').astype('Int64')
data['P.BONUS'] = data['P.BONUS'].str.replace(',', '').str.replace('$', '').astype('Int64')
data['ROUND PICK'] = data['ROUND PICK'].replace('-', np.nan).astype('Int64')
data['OVERALL DRAFT POSITION'] = data['OVERALL DRAFT POSITION'].replace('-', np.nan).astype('Int64')
data['SALARY'] = data['SALARY'].str.replace(',', '').str.replace('$', '').astype('Int64')
data['AAV'] = data['AAV'].str.replace(',', '').str.replace('$', '').astype('Int64')
data['CAP HIT'] = data['CAP HIT'].str.replace(',', '').str.replace('$', '').astype('Int64')
data['S.BONUS'] = data['S.BONUS'].str.replace(',', '').str.replace('$', '').astype('Int64')
data['IPP'] = data['IPP'].replace('-', np.nan).astype('float')
data['PDO'] = data['PDO'].replace('-', np.nan).astype('float')
```

```python
# Select all columns that are non-numeric from our data
non_numeric_columns = data.select_dtypes(exclude=[np.number]).columns.to_list()
non_numeric_columns
```

We have column 'YEAR' and column 'SEASON'. Delete one?

```python
# Check if 'YEAR' and 'SEASON' are identical
data['YEAR'].equals(data['SEASON'])

# Drop the 'YEAR' column
data.drop(columns=['YEAR'], inplace=True)
```

We have column 'POSITION' and column 'POS', delete one?

```python
# Check if 'POSITION' and 'POS' are identical
data['POSITION'].equals(data['POS'])

# Check the values of 'POSITION' and 'POS'
print(data['POSITION'].unique())
print(data['POS'].unique())

# Since 'POS' is much more complicated in nature, drop it
data.drop(columns=['POS'], inplace=True)
```

# Saving to disk
Whew! That was a lot of cleaning, but our data looks like it should be in good shape moving forward.

```python
# Save the merged data to a csv file
output_dir = '../../Data/entitiesResolved'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

output_file = output_dir + '/merged_data_clean.csv'
data.to_csv(output_file, index=False)
```
