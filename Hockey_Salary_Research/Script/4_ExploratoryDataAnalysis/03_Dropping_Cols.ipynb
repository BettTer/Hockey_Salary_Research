# Getting Rid of Unnecessary Columns Before EDA and Training

```python
import pandas as pd
import os

# Load in our data
filepath = '../../Data/entitiesResolved/merged_data_clean.csv'
data = pd.read_csv(filepath)

data.head(-5)
```

```python
cols_to_keep = ['POSITION', 'PLAYER', 'TEAM', 'GP', 'HANDED', 'AAV', 'SEASON', 'SALARY CAP PERCENTAGE', 'SALARY CAP']

# Remove columns with values that cause issues
nan_columns_to_drop = data.columns[data.isna().any()].tolist()

# List all columns that do not end with '/GP' or '/60'
all_columns_to_drop = [col for col in data.columns if not col.endswith('/GP') and not col.endswith('/60') and col not in cols_to_keep]

data = data.drop(columns=all_columns_to_drop)
print(data.shape, data.columns.to_list())
```

# Saving Final Changes

```python
# Save the merged data to a csv file
output_dir = '../../Data/entitiesResolved'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

output_file = output_dir + '/merged_data_final.csv'
data.to_csv(output_file, index=False)
```
